```python
import pymysql
import pandas as pd

db = pymysql.connect("localhost","root","","airline2" )

def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
query = 'SELECT * FROM airplane;'
run_query(query)
```

```python
query = 'SELECT ID, type FROM airplane;'
run_query(query)
```

```python
#LIMIT simply tells the database how many rows you want it to return.
query = 'SELECT * FROM airplane LIMIT 1;'
run_query(query)
```

```python
#query = 'ORDER BY ID ASC;'
query = '''
SELECT time, date, classtype FROM book
ORDER BY ID DESC
LIMIT 10;
'''
run_query(query)
```

```python
query = '''
SELECT * FROM book
WHERE classtype = "Business"
LIMIT 5;
'''
run_query(query)
```

```python
query = '''
SELECT COUNT(ID)
FROM book;
'''
run_query(query)
```

```python
query = '''
SELECT COUNT(ID) AS "Total Paid Bookings"
FROM book
WHERE paid = 0;
'''
run_query(query)
```

```python
query = '''
SELECT AVG(price) AS "Average Price"
FROM class;
'''
run_query(query)
```

```python
#GROUP BY separates rows into groups based on the contents of a particular column 
#and allows us to perform aggregate functions on each group.

query = '''
SELECT number, name 
FROM class
GROUP BY name;
'''
run_query(query)
```

```python
#Joins
#(INNER) JOIN: Returns records that have matching values in both tables.
#LEFT (OUTER) JOIN: Returns all records from the left table, and the matched records from the right table.
#RIGHT (OUTER) JOIN: Returns all records from the right table, and the matched records from the left table.
```

```python
#SELECT passanger.username, book.classtype
query = '''
SELECT *
FROM passanger
INNER JOIN book
ON passanger.username = book.username
LIMIT 10;
'''
run_query(query)
```

```python
query = '''
SELECT passanger.username, book.classtype
FROM passanger
LEFT JOIN book
ON passanger.username = book.username
LIMIT 5;
'''
run_query(query)
```
